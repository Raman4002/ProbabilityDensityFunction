```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset("iris")
```

```python
df.head(2)
```

```python
plt.figure(figsize=(3,3))
plt.hist(df.petal_length,bins=10);
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(data=df,x='petal_length',hue='species')
sns.ecdfplot(data=df,x='petal_length',hue='species')
y_ticks =np.arange(0, 1.1, 0.05)
x_ticks =np.arange(0, 8, 0.5)
plt.yticks(y_ticks)
plt.xticks(x_ticks);
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(data=df,x='sepal_length',hue='species')
sns.ecdfplot(data=df,x='sepal_length',hue='species')
y_ticks =np.arange(0, 1.1, 0.05)
plt.yticks(y_ticks);
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(data=df,x='sepal_width',hue='species')
sns.ecdfplot(data=df,x='sepal_width',hue='species')
y_ticks =np.arange(0, 1.1, 0.05)
plt.yticks(y_ticks);
```
